```python
import numpy as np
import matplotlib.pyplot as plt
import os


%matplotlib inline


```

```python
logs = {}
for filename in os.listdir('./'):
    if filename.endswith(".csv"):
        logs[filename.split('_')[1]] = np.genfromtxt(filename,delimiter=",")
    
    
    
fig = plt.figure()

fig.set_figheight(15)
fig.set_figwidth(15)

    
i = 1
for key,val in logs.items():
    try:
        ax = fig.add_subplot(220 + i)
        ax.plot(list(range(5000)),val[:5000,0])
        ax.plot(list(range(5000)),val[:5000,1])
        #ax.set_ylim([0,100])
        ax.set_title(key)
    except:
        print(len(val))
    i += 1

plt.show()
```
