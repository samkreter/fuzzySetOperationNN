```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 

noreg_add = np.genfromtxt('noreg_add_log.csv', delimiter=',')
noreg_com = np.genfromtxt('noreg_combined_log.csv', delimiter=',')
noreg_sub = np.genfromtxt('noreg_sub_log.csv', delimiter=',')

reg_add = np.genfromtxt('reg_add_log.csv', delimiter=',')
reg_com = np.genfromtxt('reg_combined_log.csv', delimiter=',')
reg_sub = np.genfromtxt('reg_sub_log.csv', delimiter=',')



print(noreg_add[-1])
print(noreg_com[-1])
print(noreg_sub[-1])

print(reg_add[-1])
print(reg_com[-1])
print(reg_sub[-1])


```

```python
plt.plot(list(range(5000)),noreg_add[:,0])
plt.plot(list(range(5000)),noreg_add[:,1])
plt.ylim([0,100])
plt.show()
```

```python
fig = plt.figure()

fig.set_figheight(15)
fig.set_figwidth(15)

ax1 = fig.add_subplot(231)
ax1.plot(list(range(5000)),noreg_add[:,0])
ax1.plot(list(range(5000)),noreg_add[:,1])
ax1.set_ylim([0,100])
ax1.set_title("Noreg Add")

ax2 = fig.add_subplot(232)
ax2.plot(list(range(5000)),noreg_sub[:,0])
ax2.plot(list(range(5000)),noreg_sub[:,1])
ax2.set_ylim([0,100])
ax2.set_title("Noreg Sub")

ax3 = fig.add_subplot(233)
ax3.plot(list(range(5000)),noreg_com[:,0])
ax3.plot(list(range(5000)),noreg_com[:,1])
ax3.set_ylim([0,100])
ax3.set_title("Noreg Com")


ax4 = fig.add_subplot(234)
ax4.plot(list(range(5000)),reg_add[:,0])
ax4.plot(list(range(5000)),reg_add[:,1])
ax4.set_ylim([0,100])
ax4.set_title("Reg Add")

ax5 = fig.add_subplot(235)
ax5.plot(list(range(5000)),reg_sub[:,0])
ax5.plot(list(range(5000)),reg_sub[:,1])
ax5.set_ylim([0,100])
ax5.set_title("Reg Sub")

ax6 = fig.add_subplot(236)
ax6.plot(list(range(5000)),reg_com[:,0])
ax6.plot(list(range(5000)),reg_com[:,1])
ax6.set_ylim([0,100])
ax6.set_title("Reg Com")


fig.savefig("test.png")
```
